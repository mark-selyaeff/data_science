# Solution

#### Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

#### Generate

```python
size = 1000 # Размер выборок
```

```python
a,b = 0.5, 0.5

unknown_rv           = sts.beta(a,b) # Бета-распределение
unknown_distribution = unknown_rv.rvs(size=size)

df = pd.DataFrame(unknown_distribution, columns=['KDE'])
ax = df.plot(kind='density')

x = np.linspace(0, 1, 100)
plt.plot(x, unknown_rv.pdf(x), label="teoretical distribution", alpha=0.5, color='black')
plt.hist(unknown_distribution, normed=True, color='blue', label='actual destribution')

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
def repeat_generation(n):
    
    # Бета-распределение
    a,b = 0.5, 0.5
    unknown_rv        = sts.beta(a,b) 
    mean_distribution = [np.mean(unknown_rv.rvs(size=n)) for i in range(size)]
    
    # Гисторграма 
    plt.hist(mean_distribution, normed=True, color='blue', label='actual destribution', bins=10)  
    
    
    # Нормальное распределение
    MU = a/(a+b)
    SIGMA = math.sqrt(((a*b)/(((a+b)**2)*(a+b+1)))/n)
    
    print 'Дисперсия нормального распределения', SIGMA
    print 'Мат. ожидание нормального распределения', MU
    
    norm_mean_rv = sts.norm(MU, SIGMA) # Нормальное распередление к которому приближаються выборочные среднее
    
    # График теоретичного плостности нормального распеределение
    x = np.linspace(0, 1, 100)
    plt.plot(x, norm_mean_rv.pdf(x), label="teoretical distribution", color='orange')
    
    
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
repeat_generation(5)
```

```python
repeat_generation(10)
```

```python
repeat_generation(50)
```

```python
repeat_generation(500)
```

1. Центральная придельна теорема работает. При маленьких размерах выборки на графиках видно приближение
2. Увеличение `n` увеличивает точность апроксимирования распределения выборочных средних нормальным
